  <h1>Listas en Python</h1>
</div>

*  Una lista es una estructura de datos que contiene conjuntos ordenados de datos entre corchetes [] separados por una coma.

  <h1>Tuplas en Python</h1>
</div>

*  Una tupla es una secuencia ordenada de elementos.
*  Se define encerrando los elementos entre paréntesis ( ), separados por una coma.
*  Se utiliza para datos fijos
*  Es inmutable: una vez definida, no se pueden eliminar, agregar ni editar los valores que contiene.
*  Una tupla se utiliza en situaciones en las que es seguro que el (conjunto de) valores que pertenecen a alguna declaración o función definida por el usuario no cambiará.
*  El acceso a los elementos de una tupla es más rápido que el de una lista.

```
primos = (2, 3, 5, 7, 11, 13)
print(type(primos))
```

```
planetas = ('mercurio', 'venus', 'tierra', 'marte', 'jupiter',
           'saturno', 'urano', 'neptuno')
print(type(planetas))
```

  <h1>Conjuntos en Python</h1>
</div>

*  Una secuencia usada para almacenar datos no-duplicados en Python.
*  Los datos no están ordenados.
*  Mutable.

```
fibo = {1, 1, 2, 3, 5}
print(type(fibo))
print("Set fibo:", fibo)
```

```
fibo[0]   # NO POSIBLE - Datos están desordenados!
```

```
primos_repetidos = [1, 1, 2, 2, 3, 3, 5, 5]
primos = set(primos_repetidos)
print("Set primes:", primos)
```

```
# 1. Crear dos conjuntos con datos de personas, una de edades y otra de pesos cada una con 5 elementos

edades = [35, 35, 40, 44, 56]
pesos = [60, 68, 63, 68, 80]
display(edades, pesos)

Cedades = set(edades)
Cpesos = set(pesos)
display(Cedades, Cpesos)

# 2. Representar con una nuevo conjuntos, un conjunto de conjuntos con (edad,peso) de cada uno de las 5 personas

personas = set(zip(Cedades, Cpesos))
display(personas)
```

  <h1>Diccionarios en Python</h1>
</div>

Un diccionario en Python es una colección de valores de claves que se utilizan para almacenar valores de datos como un mapa, que, a diferencia de otros tipos de datos que contienen solo un valor único como elemento.
Algunas características particulares son:

*  Un diccionario es una estructura de datos asociativa de longitud variable.
*  Un diccionario consiste en una colección ordenada de pares clave:valor separados por comas dentro de llaves {}.
*  Cada par clave-valor asigna la clave a su valor asociado.
*  Una clave debe ser única y un objeto inmutable.
*  Un diccionario es mutable.

```
# Creando un diccionario
my_dict = {'name': 'Jack', 'age': 26}
my_dict
```

```
my_dict['age']
```

```
# Creando un diccionario
d4 = dict([('name','nitish'),('age',32),(3,3)])
d4
```

```
# Agregando nuevos elementos al diccionario
d4['gender'] = 'male'
d4
d4['weight'] = 72
d4
```

```
# 1. Dado el siguiente diccionario de palabras + puntajes individuales, escriba un programa para ingresar y buscar el valor numérico de letra por letra
# de una palabra y despues calcular la suma de todos los valores de cada letra para cualquier palabra.
# Por ejemplo, 'Python' obtiene un puntaje de 14.

puntajesLetras = {'a': 1, 'b': 3, 'c': 3, 'd': 2, 'e': 1, 'f': 4, 'g': 2, 'h': 4, 'i': 1, 'j': 8, 'k': 5, 'l': 1, 'm': 3, 'n': 1, 'o': 1, 'p': 3, 'q': 10, 'r': 1, 's': 1, 't': 1, 'u': 1, 'v': 4, 'w': 4, 'x': 8, 'y': 4, 'z': 10}

palabra = input("Ingrese la palabra para conocer el puntaje: ")
print("\n")
puntajePalabra = 0

for letra in palabra:
  if letra in puntajesLetras:
    puntajePalabra += puntajesLetras[letra]
    print("Letra", letra, " = ", puntajesLetras[letra])
  else:
    print("Letra", letra, " no tiene puntaje asignado")


print("\nPuntaje de", palabra, " = ", puntajePalabra)
```
