```
# Elevar a la potencia array_b el array_a (array_a)^array_b

import numpy as np

array_a = np.array([1, 2, 3])
array_b = np.array([4, 5, 6])

# Método 1: Por operacioens matriciales

b2 = array_b ** 2
display(b2)
a2 = array_a ** array_b
display(a2)

'''
# Método 2: Función propia: Generar una lista con la operación de cada elemento de la base por cada elemento elevado del exponente
'''

base = []
exponente = []
resultado = []

numeros = int(input("\n\nCuántos números componen la lista?: "))

for n in range(numeros):
  base.append(float(input("Ingrese el número de la base: ")))
  exponente.append(float(input("Ingrese el número del exponente: ")))
  resultado.append(base[n] ** exponente[n])

display(resultado)
```

```
# Ordenar los elementos del array_a de menor a mayor para obtener el array_ordenado = np.array([[-4, -2, 1],[3, 5, 6]])

import numpy as np

array_a = np.array([[-4, 5, 6],[1, -2, 3]])
display(array_a)

arr_ordenado = np.sort(array_a, axis=None).reshape(array_a.shape)
display(arr_ordenado)
```

```
# El indice de masa corporal de una persona se calcula como: IMC = Peso (en Kg) / Estatura  al cuadrado (en metros).
# https://www.minsalud.gov.co/salud/Paginas/Evalue-su-peso.aspx
# Crear un vector con los pesos de 5 personas y otro vector con las 5 estaturas, calcular el IMC

estaturas = np.array([1.66, 1.67, 1.68, 1.64, 1.66])
pesos = np.array([80.0, 67.0, 78.0, 80.0, 77.0])
IMC = pesos / estaturas ** 2

print("IMC = ", IMC)
```
