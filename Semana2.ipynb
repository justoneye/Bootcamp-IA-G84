```python
# CONDICIONAL if
# Crear un programa que le pregunte la edad a una persona si es mayor o menor de edad

def edad():
    edad = int(input("Ingrese su edad: "))

    if edad>=18:
        print("La persona es MAYOR de edad")
    if edad<18:
        print("La persona es MENOR de edad")
```

```python
#CONDICIONALES else y elif
'''
Crear un programa que, dependiendo del IMC de una persona indique:
- Bajísimo: Menor a 10
- Bajo: 10 a 17
- Normal: 17 a 35
- Riesgoso: Mayor de 35
'''

peso = float(input("Ingrese su peso (en kilos): "))
altura = float(input("Ingrese su altura (en metros): "))
IMC = peso/(altura**2)

if IMC<10:
    mensaje = "Bajísimo"
elif IMC>=10 and IMC<17:
    mensaje = "Bajo"
elif IMC>=17 and IMC<35:
    mensaje = "Normal"
else:
    mensaje = "Riesgoso"

print("\n\nSu IMC es ", IMC, "(",mensaje,")")
```

```python
#CONDICIONAL while
'''
Crear un programa que, si el usuario presiona un número:
- 1: Si el usuario es mayor de edad o no
- 2: Si la persona es alta (mayor a 1.70m) o no.
- 3: Salir.
'''

print("MENÚ \n\n"+
          "   1: Si el usuario es mayor de edad o no\n"+
          "   2: Si la persona es alta (mayor a 1.70m) o no.\n"+
          "   3: SALIR.\n")
opcion = ""

while (opcion!="3"):
    opcion = input("Ingrese la opción correspondiente: ")

    if opcion=="1":
        edad()
        print("\n\n")

    elif opcion=="2":
        altura = float(input("Ingrese su altura (en metros): "))
        if altura>=1.70:
            print("La persona es ALTA")
        else:
            print("La persona es BAJA")
        print("\n\n")

    elif opcion=="3":
        print("--- FIN ---")
    else:
        print("Opción incorrecta, ingresa una opción válida nuevamente.")
```

```python
'''
Elaborar un código que muestre para un número cualquiera si es primo o no
'''
numero = int(input("Ingrese el número: "))
primo = True
divisores = 0

# Método 1: Bandera en booleano para números primos

for i in range(2, numero-1):
    if (numero % i == 0 and primo == True):
        print("El numero", numero, "es divisible por", i, ", NO es primo")
        primo = False
if (primo == True):
    print("El numero", numero, "es PRIMO")



# Método 2: Contar los divisores de un número, si existen

def primo(num):
  for i in range(2, numero-1):
      if (num % i == 0 and divisores == 0):
          print("El numero", num, "NO es PRIMO")
          divisores+=1

  if (divisores == 0):
      print("El numero", num, "es PRIMO")
```

```python
#CONDICIONAL for
'''
Elaborar un código que muestre los números impares desde el 1 hasta el número que dió el usuario
'''
numero = int(input("Ingrese el número: "))

# Método 1: FOR con 3 condiciones

for i in range(1, numero+1, 2):
    print(i)

# Método 2: Manual

for i in range(1, numero+1):
    if (i % 2 != 0):
        print(i)
```

```python
# FUNCIONES

'''
Elaborar una función que sume dos números enteros
'''


def sumar(num1, num2):
    return num1+num2

def restar(num1, num2):
    return num1-num2

def multiplicar(num1, num2):
    return num1*num2

def dividir(num1, num2):
    return num1/num2

def main():
    numero1 = float(input("Ingrese numero 1: "))
    numero2 = float(input("Ingrese numero 2: "))

    print("\n\n", "MENÚ \n   1: Sumar\n   2: Restar\n   2: Multiplicar\n   2: Dividir\n   5: SALIR.\n")
    opcion = ""

    while (opcion!="5"):
        opcion = input("Ingrese la opción correspondiente: ")

        if opcion=="1":
            print("Resultado =", sumar(numero1,numero2))

        elif opcion=="2":
            print("Resultado =", restar(numero1,numero2))

        elif opcion=="3":
            print("Resultado =", multiplicar(numero1,numero2))

        elif opcion=="4":
            print("Resultado =", dividir(numero1,numero2))

        elif opcion=="5":
            print("--- FIN ---")
        else:
            print("Opción incorrecta, ingresa una opción válida nuevamente.")

main()
```

```python
'''
Saber si un número es par o impar
'''

# Método 1: Propio

def par():
  numero = int(input("Ingrese numero para saber si es par o impar: "))

  if (numero % 2 == False):
    return "El número es PAR"
  else:
    return "El número es IMPAR"

par()
```

```python
'''
Saber si un número es par o impar
'''

# Método 2: Usando Booleanos

def par(num):

  if (num % 2 == 0):
    return True
  return False

def main():
  numero = int(input("Ingrese numero para saber si es par o impar: "))
  par(numero)

  if (numero == True):
    print ("El número es PAR")
  else:
    print ("El número es IMPAR")

main()
```
