La industria del vino dentro de sus estándares requiere la medición de una pruebas fisicoquímicas para validar la calidad del vino. Para está actividad se usarán dos conjuntos de datos relacionados con muestras de vino verde tinto y blanco del norte de Portugal.

## Reto 1: Lectura base de datos

El ejercicio propuesto es el siguiente:

```python
# 1. Cargar la base de datos de calidad de vinos de la siguiente URL
# https://archive.ics.uci.edu/static/public/186/data.csv

import pandas as pd

wines = pd.read_csv('https://archive.ics.uci.edu/static/public/186/data.csv')
display(wines)


# 2. Hacer un subconjunto con los registros con 'quality' igual a 5, 6, y 7.
quality = wines[(wines['quality'] >= 5) & (wines['quality'] <= 7)]


# 3. Imprimir el tamaño del nuevo dataframe
display(quality)
quality.size
```

## Reto 2: Clasificador de Vinos

 El ejercicio propuesto requiere el diseño de un MultiLayerPerceptron como clasificador siguiendo los siguientes pasos:

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# 4. Definir X con las características desde 0-10
nuevo = wines[(wines['quality'] >= 5) & (wines['quality'] <= 6)]
X = nuevo.iloc[:, 0:11]
display(X)

# 5. Definir y con la característica 'quality'
y = nuevo['quality']
display(y.head())

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# 6. Entrenar un MLPClassifier con los siguientes parámetros: hidden_layer_sizes=(10,10), random_state=10, activation='logistic'.
# Crear una instancia de MLPClassifier con los parámetros especificados

nlp = MLPClassifier(hidden_layer_sizes=(10,10), random_state=10, activation='logistic', max_iter=200)

# Entrenar modelo
nlp.fit(X_train, y_train)

print("----- MLPClassifier entrenado con éxito -----")


# 7. Calcular y reportar el score obtenido
score = nlp.score(X_test, y_test)
print(f"Exactitud del MLPClasiffier en el conjunto de prueba (%): {score}")
```

# Reto 3: Estimador de características

El ejercicio propuesto requiere el diseño de un MultiLayerPerceptron como regresor siguiendo los siguientes pasos:

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

# 8. Definir X con las características desde 0-9

X = nuevo.iloc[:, 0:10]
display(X)

# 9. Definir y con la característica 'alcohol'
y = nuevo['alcohol']
display(y.head())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 10. Entrenar un MLPRegressor con los siguientes parámetros: hidden_layer_sizes=(100), random_state=10, activation='tanh'

nlp_regressor = MLPRegressor(hidden_layer_sizes=(100), random_state=10, activation='tanh', max_iter=200)
nlp_regressor.fit(X_train, y_train)

print("----- MLPRegressor entrenado con éxito -----")

# 11. Realizar la predicción y reportar el 'mean square error" obtenido!

score = nlp.score(X_test, y_test)
print(f"Exactitud de MLPRegressor (%): {score}")

y_pred = nlp_regressor.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print(f"Error cuadrático medio (%): {mse}")
```
