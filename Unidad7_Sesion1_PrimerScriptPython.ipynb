# NumPy

NumPy es la biblioteca más popular de Python y sirve para almacenar vectores y matrices de números para poder realizar cálculos con ellas. Esto no solo permite escribir código a menudo más conciso, sino que también hace que el código sea más rápido, ya que la mayoría de las rutinas de NumPy se implementan en C para mayor velocidad.

```
# Debemos importar Numpy antes de poder usarla
import numpy as np
```

```
# Para crear un vector y una matriz se debe usar np.array a una lista
vector = np.array([1, 2, 3])
matriz = np.array([[1, 2, 3],[4, 5, 6]])
vector, matriz
```

```
# Buscando el elemnto con posición fila= 1, columna=2
matriz[1, 2]
```

```
matriz[1]
```

```
matriz[:, 2]
```

```
# Dimension del "array"
print(vector.ndim)
print(matriz.ndim)
```

```
# Tamaño del "array"
print(vector.shape)
print(matriz.shape)
```

```
# El tipo de datos del "array"
vector.dtype, matriz.dtype
```

```
# Crear un array de 4x3 y usar .reshape para generar: un array de 3x4 y otro array de 6x2

import numpy as np

arreglo = np.array([[2, 20, 24],[7, 17, 23],[12, 18, 43],[18, 23, 43]])
print("\nArreglo:\n", arreglo, "\nShape:", arreglo.shape)

nuevoArreglo = arreglo.reshape(3, 4)
print("\nNuevo Arreglo:\n", nuevoArreglo, "\nShape:", nuevoArreglo.shape)


# Métodos 2: Números al azar

from numpy import random

print("\n\nMétodos 2: Números al azar ----------")

arregloRandom = random.randint(50, size=(4, 3))
print("\nArreglo Random:\n", arregloRandom, "\nShape:", arregloRandom.shape)

nuevoRandom = arregloRandom.reshape(3, 4)
print("\nNuevo Random:\n", nuevoRandom, "\nShape:", nuevoRandom.shape)
```

```
# Crear una lista de la estatura de n personas.

import numpy as np

listaEstaturas = []
personas = int(input("Cuántas personas van a añadir a lista?: "))

for i in range(personas):
  estatura = float(input(" - Ingrese estatura: "))
  listaEstaturas.append(estatura);

arrayLista = np.array(listaEstaturas)

print("\n Media:", arrayLista.mean())
```
