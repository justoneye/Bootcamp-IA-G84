```python
frutas = ["banano", "manzana", "mango", "pera"]
datos=[1,2,3,4,5]

# Función lenght para saber el largo de un arreglo

print(len(frutas))
print(len(datos))

# Imprimir el dato de la posición [i] del arreglo frutas

print("\nFrutas[]")
print(len(frutas[0]))
print(len(frutas[1]))
print(len(frutas[2]))
print(len(frutas[3]))

# Sentencia for para imprimir arreglo frutas

print("\nContenido de array frutas[]")
for i in range(len(frutas)):
  print(frutas[i])

# Sentencia for para imprimir arreglo de datos

print("\nContenido de array datos[]")
for dato in range(len(datos)):
  print(datos[dato])
```

```python
'''
Calcular el área de un círculo pidiendo al usurio el radio
'''

import math

def areaCirculo(r):
  circunferencia = math.pi * r ** 2
  mensaje = print("El área del círculo es: " + str(circunferencia))
  return mensaje

radio = float(input("Ingrese el radio: "))
areaCirculo(radio)
```
