```python
# Función para número par
def par(num):
  if (num % 2 == False):
    return "El número es PAR"
  else:
    return "El número es IMPAR"



# Función para número primo
def primo(num):
  primo = True
  divisores = 0
  respuesta = ""

  for i in range(2, num-1):
      if (num % i == 0 and divisores == 0):
          respuesta = "El numero NO es PRIMO"
          divisores+=1

  if (divisores == 0):
      respuesta = "El numero es PRIMO"

  return respuesta



# Función para ejecutar el menú inicial
def main():
    numero = int(input("Ingrese número: "))
    print("\n\n", "MENÚ: Saber si un número es:\n   1: Par o Impar\n   2: Primo\n   3: SALIR.\n")
    opcion = ""

    while (opcion!="3"):
        opcion = input("Ingrese la opción correspondiente: ")

        if opcion=="1":
            print(par(numero), "\n")

        elif opcion=="2":
            print(primo(numero), "\n")

        elif opcion=="3":
            print("--- FIN ---")
        else:
            print("Opción incorrecta, ingresa una opción válida nuevamente.\n\n")

main()
```

```python
'''
Crear un menú con las siguientes opciones:
  1. Calcular el área de un cuadrado
  2. Calcular el área de un círculo
  3. Calcular el área de un rectángulo
  4. Salir
'''

import math

# Función para área de un cuadrado
def cuadrado(lado):
    lado = int(input("Ingrese la medida del lado del cuadrado (sin unidades): "))
    return lado ** 2


# Función para área de un círculo
def circulo(radio):
  radio = float(input("Ingrese el radio del círculo (sin unidades): "))
  return math.pi * radio ** 2


# Función para área de un rectangulo
def rectangulo(alto, ancho):
  alto = int(input("Ingrese el alto del rectángulo (sin unidades): "))
  ancho  = int(input("Ingrese el ancho del rectángulo (sin unidades): "))
  return alto * ancho



# Función para ejecutar el menú inicial
def main():
    print("\n\n", "MENÚ: \n 1. Calcular el área de un cuadrado \n 2. Calcular el área de un círculo \n 3. Calcular el área de un rectángulo 4: SALIR.\n")
    opcion = ""
    alto = ""
    ancho = ""

    while (opcion!="4"):
        opcion = input("Ingrese la opción correspondiente: ")

        if opcion=="1":
            print(cuadrado(alto), "\n")

        elif opcion=="2":
            print(circulo(alto), "\n")

        elif opcion=="3":
            print(rectangulo(alto, ancho), "\n")

        elif opcion=="4":
            print("--- FIN ---")
        else:
            print("Opción incorrecta, ingresa una opción válida nuevamente.\n\n")

main()
```
