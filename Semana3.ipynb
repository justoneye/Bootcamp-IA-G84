```python
# Función para número par
def par(num):
  if (num % 2 == False):
    return "El número es PAR"
  else:
    return "El número es IMPAR"



# Función para número primo
def primo(num):
  primo = True
  divisores = 0
  respuesta = ""

  for i in range(2, num-1):
      if (num % i == 0 and divisores == 0):
          respuesta = "El numero NO es PRIMO"
          divisores+=1

  if (divisores == 0):
      respuesta = "El numero es PRIMO"

  return respuesta



# Función para ejecutar el menú inicial
def main():
    numero = int(input("Ingrese número: "))
    print("\n\n", "MENÚ: Saber si un número es:\n   1: Par o Impar\n   2: Primo\n   3: SALIR.\n")
    opcion = ""

    while (opcion!="3"):
        opcion = input("Ingrese la opción correspondiente: ")

        if opcion=="1":
            print(par(numero), "\n")

        elif opcion=="2":
            print(primo(numero), "\n")

        elif opcion=="3":
            print("--- FIN ---")
        else:
            print("Opción incorrecta, ingresa una opción válida nuevamente.\n\n")

main()
```

```python
'''
Crear un programa que:
1. Adicionar una venta
2. Mostrar el promedio de ventas
3. Salir
'''

def main():
  menu = "\n\nMENÚ: \n 1. Adicionar una venta. \n 2. Mostrar el promedio de ventas. \n 3. Salir del programa.\n\nPor favor ingrese la opción que desee:"
  opc = ""
  ventas = []

  while opc!="3":
    print(menu)
    opc = input()

    if opc=="1":
      venta = float(input("Ingrese la nueva venta: $"))
      ventas.append(venta)
      print("\nNueva venta adicionada. En total hay", len(ventas), "ventas registradas.")

    elif opc=="2":
      sumaVentas = 0.0

      for i in range (len(ventas)):
        sumaVentas = sumaVentas + ventas[i]
      sumaVentas = sumaVentas / len(ventas)
      print("El promedio de las ventas es: $", sumaVentas, "\n")
    elif opc=="3":
      print("--- Programa finalizado. Gracias.")
    else:
      print("La opción es incorrecta, por favor inténtelo nuevamente\n\n")

main()
```

```python
'''
Crear un programa que:
1. Adicionar una venta
2. Mostrar el promedio de ventas
3. Salir
'''

# Ejercicio anterior aplicando funciones creadas apra el manejo de arreglos

def agregarElemento(elemento, conjunto):
  conjunto.append(elemento)



def calcularPromedio(conjunto):
  suma = 0.0

  for i in range (len(conjunto)):
    suma = suma + conjunto[i]
  promedio = suma / len(conjunto)
  mensaje = print("El promedio de las ventas es: $", suma, "\n")

  return mensaje



def main():
  menu = "\n\nMENÚ: \n 1. Adicionar una venta. \n 2. Mostrar el promedio de ventas. \n 3. Salir del programa.\n\nPor favor ingrese la opción que desee:"
  opc = ""
  ventas = []
  mediosDePago = []

  while opc!="3":
    print(menu)
    opc = input()

    if opc=="1":
      venta = float(input("\nIngrese la nueva venta: $"))
      agregarElemento(venta, ventas)
      print("Nueva venta adicionada. En total hay", len(ventas), "ventas registradas.")

      medio = input("\nIngrese medio de pago: ")
      agregarElemento(medio, mediosDePago)
      print("Medio de pago (", medio, ") agregado")

    elif opc=="2":
      if len(ventas)!=0:
        calcularPromedio(ventas)
      else:
        print("No hay ventas registradas.")
    elif opc=="3":
      print("--- Programa finalizado. Gracias.")
    else:
      print("La opción es incorrecta, por favor inténtelo nuevamente\n\n")

main()
```
