```python
# Función para número par
def par(num):
  if (num % 2 == False):
    return "El número es PAR"
  else:
    return "El número es IMPAR"



# Función para número primo
def primo(num):
  primo = True
  divisores = 0
  respuesta = ""

  for i in range(2, num-1):
      if (num % i == 0 and divisores == 0):
          respuesta = "El numero NO es PRIMO"
          divisores+=1

  if (divisores == 0):
      respuesta = "El numero es PRIMO"

  return respuesta



# Función para ejecutar
def main():
    numero = int(input("Ingrese número: "))
    print("\n\n", "MENÚ: Saber si un número es:\n   1: Par o Impar\n   2: Primo\n   3: SALIR.\n")
    opcion = ""

    while (opcion!="3"):
        opcion = input("Ingrese la opción correspondiente: ")

        if opcion=="1":
            print(par(numero), "\n")

        elif opcion=="2":
            print(primo(numero), "\n")

        elif opcion=="3":
            print("--- FIN ---")
        else:
            print("Opción incorrecta, ingresa una opción válida nuevamente.\n\n")

main()
```
