```
# Definir una función para calcular el IMC donde se ingrese edad y tamaño.
# Sin embargo, si la edad ingresada es menor a 19, debe imprimir un mensaje alertando que NO se puede calcular el IMC.

def calcularIMC(edad, altura, peso):
  IMC = peso/(altura**2)

  if (edad <= 19):
    mensaje = print("\nNO se puede calcular el IMC")
  else:
    mensaje = print("\nEl IMC es", IMC)

  return mensaje



edadUsuario = int(input("Ingrese su edad: "))
pesoUsuario = float(input("Ingrese su peso (en kilos): "))
alturaUsuario = float(input("Ingrese su altura (en metros): "))

calcularIMC(edadUsuario, pesoUsuario, alturaUsuario)
```

```
# Ingrese con 'input' un valor positivo.
# Escribir una función para calcular el número factorial del número ingresado.
# https://es.wikipedia.org/wiki/Factorial

def factorial(numero):
  factorial = 1
  for i in numero:
    factorial*=i

  return factorial


num = int(input("Ingrese el número: "))
print("Factorial de", num, "es", factorial(num))
```

```
# Definir una función para calcular el IVA que necesite como argumentos el valor del objeto y un texto con la siguiente condición:
# Se debe ingresar el valor de 1 ítem a comprar, e ingresar una cadena de texto, donde si se ingresa "NO IVA" se aplica un IVA del 16% y si se ingresa "IVA", se aplica del 19%.
# Finalmente imprima el valor del objeto, el valor del IVA y un texto que me diga el porcentaje del IVA aplicado.

def calcularIVA(valor, mensaje):
  porcentaje = 0.0

  if mensaje == "NO IVA":
    porcentaje = 1.16
  elif mensaje == "IVA":
    porcentaje = 1.19

  return("IVA =", valor*porcentaje)

valor = float(input("Ingrese el valor del producto: "))
mensaje = input("Observaciones: ")
calcularIVA(valor, mensaje)
```

```
# Definir dos funciones: 1 para convertir de pulgadas a cm y 1 de centimetro a pulgadas.
# Haga prueba con valores conocidos en cm o pulgadas

def inCM(medida):
  cm = medida * 2.54
  return (print(medida, "in =", cm, "cm"))

def CMin(medida):
  inc = medida * 0.393701
  return (print(medida, "cm =", inc, "in"))

def main():
  convertir = float(input("Ingrese las unidades a convertir: "))
  opc = ""

  while opc!="0":
    opc = input("\n\nMENÚ: \n 1. Pulgadas (in) a Centímetros (cm) \n 2. Centímetros (cm) a Pulgadas (in) \n 0. Salir del programa.\n\nPor favor ingrese la opción que desee: ")

    if opc=="1":
      inCM(convertir)

    elif opc=="2":
      CMin(convertir)

    else:
      print("Opción incorrecta, inténtelo nuevamente\n\n")

main()
```
