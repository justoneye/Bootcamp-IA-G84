```python
# Pedirle al usuario 3 números. Dar la suma, multiplicacion y decir cuál es el menor.

numeros = []

a = float(input("Ingrese el número correspondiente: "))
numeros.append(a)
b = float(input("Ingrese el número correspondiente: "))
numeros.append(b)
c = float(input("Ingrese el número correspondiente: "))
numeros.append(c)

suma = a+b+c
multiplicacion = a*b*c

menor = 0.0

for i in range(len(numeros)):
  if (i == 0 or numeros[i] < menor):
    menor = numeros[i]

print("\n\nSuma: ", suma, "\nMultiplicacion", multiplicacion, "\nNúmero menor:", menor)
```
