Las cadenas se utilizan para almacenar texto. Se pueden delimitar mediante comillas simples o dobles.

```
cadena1 = "Hola Mundo"
cadena2 = 'Hola a todos'
print(cadena1, cadena2)
```

Podemos acceder a un elemento o a un conjunto de elementos de una cadena:

```
cadena1[3]
```

```
cadena1[5:]
```

Con el operador +, podemos unir o concatenar dos o más cadenas:

```
cadena1 + " , " + cadena2
```

## Condicionales

Los condicionales son una forma de ejecutar código dependiendo de si una condición es Verdadera o Falso. Al igual que en otros lenguajes, Python admite if y else, pero else if se contrae a elif, como lo demuestra el siguiente ejemplo.

```
# Correr la siguiente linea 3 veces, con un número positivo, otro negativo y con 0
numero = int(input('Ingrese un número entre -10 y 10'))
if numero < 0:
  print("negativo")
elif numero == 0:
  print("nulo")
else: # numero > 0
  print("positivo")
```

## Bucles

Los bucles son una forma de ejecutar un bloque de código varias veces. Hay dos tipos principales de bucles: bucles while y bucles for.

```
lista = [1, 2, 3]
```

Usando while

```
i = 0
while i < len(lista):
  print(lista[i])
  i += 1 # equivalente a i = i + 1
```

Usando for

```
for i in range(len(lista)):
  print(lista[i])
```

De otra forma:

```
for element in lista:
  print(element)
```

```
# Usar un bucle for que calcule la norma euclidiana de un vector, representada como una lista.
# https://en.wikipedia.org/wiki/Norm_(mathematics

vector = [0.5, -1.2, 3.3, 4.5]
vector_2 = []
sum_vect = 0

for i in range(len(vector)):
  vector_2.append(vector[i] ** 2)
  sum_vect += vector_2[i]

print("Norma euclidiana: ", sum_vect ** 0.5)
```

```
# Calcular la norma euclidiana de un vector según los valores uqe da el usuario
import math

vector = []
n = int(input("¿Cuántos valores tendrá el vector? "))
sum_vect = 0.0

for i in range(n):
  vector.append(float(input("Ingrese el valor: ")))
  sum_vect += math.pow(vector[i], 2)

print("Norma euclidiana:", math.sqrt(sum_vect))
```

```
# Usar 2 bucles for para encontrar el valor máximo de una matriz

import numpy as np

matriz = np.array([[-4, -2, 1],[3, 5, 6]])
maximo = 0.0

for i in range(2):
  for j in range(3):
    if (maximo < matriz[i][j]):
      maximo = matriz[i][j]

print("Valor máximo de la matríz: ", maximo)
```
