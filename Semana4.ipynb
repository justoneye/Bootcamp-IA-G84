```python
'''
Simulación de manejo de datos de las ventas B2C de productos alimenticios
'''
# Método 1: Método propio

def main():
  opc = ""
  ventas = []

  while (opc != "5"):
    opc = input("VENTAS B2C - MENÚ:\n\n 1. Adicionar una venta\n 2. Buscar el máximo\n 3. Buscar el mínimo\n 4. Calcular media, moda y desviación estándar\n 5. SALIR\n\nIngrese opción: ")

    # Opción 1 ----------
    if opc == "1":
      venta = float(input("Ingrese la venta del día: $"))
      ventas.append(venta)
      print("\n\n")


    # Opción 2 ----------
    elif opc == "2":
      mayor = 0.0
      for i in range(len(ventas)):
        if (i == 0 or ventas[i] > mayor):
          mayor = ventas[i]
      print("Venta máxima: $", mayor, "\n\n")


    # Opción 3 ----------
    elif opc == "3":
      menor = 0.0
      for i in range(len(ventas)):
        if (i == 0 or ventas[i] < menor):
          menor = ventas[i]
      print("Venta mínima: $", menor, "\n\n")

main()
```

```python
'''
Simulación de manejo de datos de las ventas B2C de productos alimenticios
'''
# Método 2: Usando Pandas

import pandas as pd

def main():
  opc = ""
  ventas = pd.Series([250000, 350000,1000000, 4500000, 250000, 250000])

  while (opc != "5"):
    opc = input("VENTAS B2C - MENÚ:\n\n 1. Adicionar una venta\n 2. Buscar el máximo\n 3. Buscar el mínimo\n 4. Calcular media, moda y desviación estándar\n 5. SALIR\n\nIngrese opción: ")

    # Opción 1 ----------
    if opc == "1":
      pass
      venta = float(input("Ingrese la venta del día: $"))
      ventas.append(venta)
      print("\n\n")


    # Opción 2 ----------
    elif opc == "2":
      mayor = ventas.max()
      print("Venta máxima: $", mayor, "\n\n")


    # Opción 3 ----------
    elif opc == "3":
      menor = ventas.min()
      print("Venta mínima: $", menor, "\n\n")

    elif opc == "4":
      media = ventas.mean()
      moda = ventas.mode()
      desviacion = ventas.std()

      print("- Media: ", media, "\n- Moda: ", moda, "\n- Desviación:", desviacion)

main()

```

```python
# Creación de DataFrame

import pandas as pd

def main():
  df = pd.DataFrame( {'nombre':["Un nombre","Dos nombres","Tres nombres","Ultimo nombre"],
                      'ventas': [100000, 250000, 500000, 50000]} )
  opc = ""

  while (opc != "5"):
    opc = input("VENTAS B2C - MENÚ:\n\n 1. Buscar el máximo\n 2. Buscar el mínimo\n 3. Calcular desviación estándar\n 4. Calcular moda\n 5. SALIR\n\nIngrese opción: ")

    # Opción 1 ----------
    if opc == "1":
      mayor = df["ventas"].max()
      print("Venta máxima: $", mayor, "\n\n")


    # Opción 2 ----------
    elif opc == "2":
      menor = df["ventas"].min()
      print("Venta mínima: $", menor, "\n\n")


    # Opción 3 ----------
    elif opc == "3":
      desviacion = df["ventas"].std()
      print("Desviación estánda: $", desviacion, "\n\n")

    elif opc == "4":
      moda = df["ventas"].mode()
      print("Moda:\n", moda, "\n\n")

    elif opc == "5":
      print("---------- SALIR ----------")

    else:
      print("Opción incorrecta, intente nuevamente")

main()
```
